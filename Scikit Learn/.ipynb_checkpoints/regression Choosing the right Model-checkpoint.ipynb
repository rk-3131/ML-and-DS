for the regression problem
Get the california house data set

* Link for the dataset https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_california_housing.html

```python
 from sklearn.datasets import fetch_california_housing
```

```python
cal_house = fetch_california_housing()
```

```python
cal_house
```

Here the data that we got is int the form of dictionary Lets convert it into dataframe

```python
import pandas as pd
```

```python
housing_df = pd.DataFrame(cal_house['data'], columns=cal_house["feature_names"])
# data has the attribute or the feature values
# feature names are the name of the features
# target has all the values corresponding to each of the row
# target names has names of the target columns
```

```python
housing_df
```

```python
housing_df["target"] = cal_house["target"]
```

```python
housing_df.head()
```

```python
housing_df.tail()
```

Now as we have to choose the model we can refer to the map that is provided there in the website : https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge

X = housing_df.drop("target", axis=1)
y = housing_df.target

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)

model = Ridge()

model.fit(X_train, y_train)

model.score(X_test, y_test)
```

Look there it has very low accuracy there hence we will now try to explore different models and see which one of them performs well with the data that we are provided with

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import LassoLars
from sklearn import svm
# lets check these three models and how they work and all

lasso = Lasso()
lasso_l = LassoLars()
# sv = svm()

lasso.fit(X_train, y_train)
lasso_l.fit(X_train, y_train)
# sv.fit(X_train, y_train)

m1_score = lasso.score(X_test, y_test)
m2_score = lasso_l.score(X_test, y_test)
# m3_score = sv.score(X_test, y_test)

m1_score, m2_score, m2_score
```

These models have performed worst here and hence we have to make selection of the new model here
After going through map lets select the randomforest algorithm

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X = housing_df.drop("target", axis=1)
y = housing_df.target

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)

model = RandomForestRegressor()

model.fit(X_train, y_train)

model.score(X_test, y_test)
```
