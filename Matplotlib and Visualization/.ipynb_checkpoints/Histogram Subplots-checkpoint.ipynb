```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
```

## Histograms

```python
x = np.random.randn(1000)

# randn will give us the data in the form of normal distribution data

# We can see that by using the concept of having the line plot

fig, ax = plt.subplots()

ax.plot(x);
```

```python
fig, ax = plt.subplots()

ax.hist(x);
```

```python
# What if we want to have multiple plots going around at the same time
# They are called as subplots
```

## There are two options for the subplots

```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(nrows=3,
                                            ncols=2
                                            )
```

```python
# Option 1
fig, ((ax1, ax2, ax7), (ax3, ax4, ax8), (ax5, ax6, ax9)) = plt.subplots(nrows=3,
                                            ncols=3
                                            )
```

```python
# For having to implement the subplots we have to use the concept of rows and columns
# Those subplots that we want to put in the same row will be present in the same parenthesis
# Hence it will make us easy to identify which ax has which of the graph
```

```python
# Now I will show you a two row and two column graph of the same data which we have learnt until now
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2
                                            )

x = [1,2,3,4,5]
y = [10,20,30,20,10]

ax1.plot(x, y)
ax2.scatter(x, y)
ax3.hist(x)
ax4.bar(x, y)
```

```python
# There is another method where we can implement the subplots 
```

```python
fig, ax = plt.subplots(nrows=2,
                      ncols=2)

ax[0,0].plot(x, y)
ax[0,1].scatter(x, y)
ax[1,0].hist(x)
ax[1,1].bar(x, y)
```
