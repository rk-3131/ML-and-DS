# Making figure with numpy array

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Create some data
x = np.linspace(1, 100, 50)
x[0:10]
```

```python
# Now just plot the data
fig, ax = plt.subplots()

ax.plot(x, x**2);

# Default it used the line plot
```

```python
# Now we will create the scatter plot
```

```python
fig, ax = plt.subplots()

ax.scatter(x, np.exp(x));
```

```python
# Lets create another one using sin wave
x = np.linspace(1, 10, 100)
fig, ax = plt.subplots()

ax.scatter(x, np.sin(x))
```

```python
# Make a plot from the dictionary directly
nut_butter_prices = {
    "Almond Butter" : 10, 
    "Peanut Butter" : 8,
    "Cashew Butter" : 12
}

fig, ax = plt.subplots()

ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())

# For the bar graph we dont have to specify the x and y explictly 
# Only thing that neeeds to be done is just making sure that the x axis stuff has to be passed first and their values must be passed later
# The value are taken as the heights and not as the y 
```

```python
x = ["RK", "Saurabh", "Piyush", "Avi", "Sush", "Nachi"]
y = [85,90,98,99,100,91]
```

```python
fig, ax = plt.subplots()

ax.bar(x, y)
```

```python
marks = {
    "Rk" : 98,
    "Saurabh" : 99,
    "Piyush" : 91,
    "Avi" : 95,
    "Nachiket" : 90,
    "Sush" : 100
}

fig, ax = plt.subplots()

ax.bar(marks.keys(), marks.values());

ax.set(
    title="Marks of students",
    xlabel="Students Name",
    ylabel="Marks"
);
```

```python
# As here we have created the bar graph which has vertical dimension we can create the bar graph which will have horizontal dimension
```

```python
fig, ax = plt.subplots()

ax.barh(list(marks.keys()), list(marks.values()));
```

```python
# One thing to remember while plotting the horizontal bar plot we have to convert the elements in the dictionary in the list and then we can use it for the purpose of having the the horizontal graph
```
