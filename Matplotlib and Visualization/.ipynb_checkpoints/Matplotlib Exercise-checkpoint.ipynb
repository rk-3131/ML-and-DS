```python
# Import pandas as pd
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view

cars_sales = pd.read_csv('car_sales.csv')
cars_sales
```

```python
# Try to plot the 'Price' column using the plot() function

# cars_sales["Price"].plott()
# It wont work as the price is in the form of string
```

```python
cars_sales["Price"] = cars_sales["Price"].str.replace("[\$\,]", "")
```

```python
cars_sales
```

```python
cars_sales["Price"] = cars_sales["Price"].astype(float).astype(int)
```

```python
cars_sales
```

```python
cars_sales["Price"].plot()
# Now we are able to do it as we have converted it into integer
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
cars_sales["Total Sales"] = np.cumsum(cars_sales["Price"])

# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
cars_sales["Sale Date"] = pd.date_range("12/01/2023", periods=(len(cars_sales)))

# View the car_sales DataFrame
cars_sales
```

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
plt.plot(cars_sales["Sale Date"], cars_sales["Total Sales"])
```

```python
# Convert the 'Price' column to the integers
# Already done

# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
plt.scatter(cars_sales["Odometer (KM)"], cars_sales["Price"])
```

```python
# Import "../data/heart-disease.csv" and save it to the variable "heart_disease"

heart_disease = pd.read_csv('90 - heart-disease.csv')
```

```python
heart_disease
```

```python
# View the first 10 rows of the heart_disease DataFrame
heart_disease.head(10)
```

```python
# Create a histogram of the "age" column with 50 bins
heart_disease["age"].plot.hist(bins=50)
```

```python
# Call plot.hist() on the heart_disease DataFrame and toggle the
# "subplots" parameter to True

heart_disease.plot.hist(subplots=True, 
                       figsize=(10,10))
```

```python
# Call the same line of code from above except change the "figsize" parameter
# to be (10, 30)

heart_disease.plot.hist(subplots = True, 
                       figsize=(10, 30))
```

```python
# heart_disease["age"]
```

```python
# Replicate the above plot in whichever way you see fit
# Note: The method below is only one way of doing it, yours might be
# slightly different

# Create DataFrame with patients over 50 years old
over_50 = heart_disease[heart_disease["age"] > 50]

# Create the plot
fig, ax = plt.subplots()

# Plot the data
age_vs_chol = ax.scatter(x=over_50['age'], 
          y=over_50['chol'], 
          c=over_50['target'])

# Customize the plot
ax.set(title="Age Vs Cholestrol", 
      xlabel="Age", 
      ylabel="Cholestrol")

# add the legend
ax.legend(*age_vs_chol.legend_elements(), title="Target")

# Add a meanline
ax.axhline(over_50["chol"].mean(), 
          linestyle="--")
fig.savefig("age vs chol.png")
```

```python
# Another plot in the same manner where we will have the plot such that age vs thalach
```

```python
# Create the fig and ax
fig, ax = plt.subplots(figsize=(10, 6))

# Creating the plot
age_vs_thalach = ax.scatter(x=over_50["age"], 
                           y = over_50["thalach"], 
                           c = over_50["target"])

# set the title xlabel and ylabel
ax.set(title="Age VS Thalach", 
      xlabel="Age", 
      ylabel="Thalach")

# Set the legends
ax.legend(*age_vs_thalach.legend_elements(), title="Target")

# Create the middle line 
ax.axhline(over_50["thalach"].mean(), 
          linestyle="--")
```

```python
# Now we will again recreate the plot from the above code and I will be able to see the changes that will be present there

# Create the plot
fig, ax = plt.subplots(figsize=(10,6))

# plot and deploy the elements onto the plot
age_vs_thalach = ax.scatter(x = over_50["age"], 
                           y = over_50["thalach"], 
                           c= over_50["target"])


# Set basic info
ax.set(title="Age vs Thalach", 
      xlabel="Age", 
      ylabel="Thalach")

# Set the legend
ax.legend(*age_vs_thalach.legend_elements(), title="Target")


# Set the horizontal line
ax.axhline(over_50["thalach"].mean(), linestyle="--")
```
