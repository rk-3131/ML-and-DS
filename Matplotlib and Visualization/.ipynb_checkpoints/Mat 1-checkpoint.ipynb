```python
# This way of plotting the element on the X axis with respect to the element on the y axis is called as stateless plotting
# There are two methods of applying and implementing the pyplot into the notebook
# One is by using the pyplot API and other one is by using the Object oriented approach
```

```python
# Method 1
fig = plt.figure() # This is used to plot the plot and store it in a variable
ax = fig.add_subplot()  # Add some axes to the plotted plot
plt.show()

# Above line of code are simillar to 
# plt.plot()
```

```python
# Second method (recommened)
fig, ax = plt.subplots()
ax.plot(x, y);
```

# MatPlotLib Example workFlow

```python
# 1. import matplotlib and get it ready for plotting in jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt

# 2. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 3. Set up plot
fig, ax = plt.subplots(figsize=(12,12))  #width and height

# 4. Plot the data
ax.plot(x, y)

# 5.Customizing the plot
ax.set(title="Sample Plot",
      xlabel="X-axis",
      ylabel="Y-axis")

# 6. Save and show (you save the whole figure)
fig.savefig("sample_plot.png")
```
