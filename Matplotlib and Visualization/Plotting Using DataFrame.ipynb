```python
cars_sales["Price"] = cars_sales["Price"].str.replace("[\$\,]", "")
```

```python
type(cars_sales["Price"][0])
```

```python
cars_sales
```

```python
cars_sales["Price"] = cars_sales["Price"].astype(float).astype(int)
# The column in the dataframe was stored as the string first
# Hence it was first converted into the float and then that float is converted into integers
```

```python
type(cars_sales["Price"][0])
```

```python
# Now lets just add the new column which will have the date where the cars were sold
cars_sales["Sold Date"] = pd.date_range("12/01/2023", periods=(len(cars_sales)))
```

```python
cars_sales
```

```python
# Now lets just add another column which will have the ammount that we have made until any particular date and then we can get the idea how much money we have made untill now
```

```python
cars_sales["Money Made"] = cars_sales["Price"].cumsum()
cars_sales
# cumsum = cummulitive sum
```

```python
 cars_sales.plot(x = "Sold Date", y = "Money Made")
```

```python
cars_sales.plot(x = "Sold Date", y = "Money Made", kind = "scatter")
# The added paramter kind can be used to get the different types of plot
```

```python
cars_sales.plot(x = "Sold Date", y = "Money Made", kind = "bar")
```

```python
cars_sales.plot(x = "Sold Date", y = "Money Made", kind = "line")
```
