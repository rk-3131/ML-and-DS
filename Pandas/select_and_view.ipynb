```python
# Lets convert our Price values which are present in the string form
# it has comma and . which in not integer
cars_data
```

```python
def convert_number(str):
    ans = "";
    for i in str:
        if i == '.':
            break
        elif i == ',':
            continue
        elif i == '$':
            continue
        else:
            ans += i
    
    return ans
            
```

```python
cars_data["Price"] = cars_data["Price"].apply(convert_number).astype(int)
```

```python
# Now we have the dataframe where we have been given missing values 
# we have to make sure that this dataframe's missing values are either dropped or they are replaced by the mean values of the datatset
# missing_dataframe["Price"].mean() now this is giving me error and still saying it is float how can this be possible
```

```python
import math

def convert_price(price_str):
    if isinstance(price_str, float) and math.isnan(price_str):
        return None  # or any other value that represents NaN in your context
    
    ans = ""
    for char in price_str:
        if char == '$' or char == ',':
            continue
        else:
            ans += char
    
    # Convert the result to an integer
    try:
        result = int(ans)
    except ValueError:
        print(f"Error converting {price_str} to integer.")
        return None  # or handle the error in a way that makes sense for your application
    
    return result
```

```python
missing_dataframe["Price"] = missing_dataframe["Price"].apply(convert_price)
```

```python
missing_dataframe
```
