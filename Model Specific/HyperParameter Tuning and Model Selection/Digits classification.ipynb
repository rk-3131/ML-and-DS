```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
file = pd.DataFrame(digits.data, columns=digits.feature_names)
file['target'] = digits.target
```

```python
file.head()
```

```python
X = file.drop('target', axis=1)
y = file.target
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(
    SVC(),
    {
        'C' : [1,10,20],
        'kernel' : ['rbf', 'linear', 'poly']
    }, 
    return_train_score=False,
    cv = 5
)
```

```python
clf.fit(X, y)
```

```python
SVC_score = pd.DataFrame(clf.cv_results_)
```

```python
clf.best_params_
```

```python
model_estimator = {
    'svm' : {
        'model' : SVC(), 
        'params' : {
            'C' : [1,10,20], 
            'kernel' : ['rbf', 'linear', 'poly']
        }
    }, 
    'random_forest': {
        'model': RandomForestClassifier(), 
        'params' : {
            'n_estimators' : [10, 100, 1000], 
            'criterion' : ["gini", "entropy", "log_loss"]
        }
    }, 
    'logistic_regression': {
        'model' : LogisticRegression(), 
        'params' : {
            'penalty' : ['l1', 'l2', 'elasticnet'], 
            'C' : [1,10,20], 
            'solver' : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
        }
    }, 
    'gaussian' : {
        'model' : GaussianNB(), 
        'params' : {
            
        }
    }, 
    'multiNB' : {
        'model' : MultinomialNB(), 
        'params' : {
            'alpha' : [1,10,20], 
            'fit_prior' : [True, False]
        }
    }, 
    'decision_tree' : {
        'model' : DecisionTreeClassifier(), 
        'params' : {
            'criterion' : ["gini", "entropy", "log_loss"], 
            'splitter' : ["best", "random"]
        }
    }
}
```

```python
model_wise_score = []

for models, info in model_estimator.items():
    clf = GridSearchCV(info['model'], info['params'], cv = 5, return_train_score=False)
    clf.fit(X, y)
    model_wise_score.append(
        {
            "name" : models, 
            'best score' : clf.best_score_, 
            'Best params' : clf.best_params_
        }
    )
```
