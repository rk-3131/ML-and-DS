```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
file = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
file["target"] = iris.target
```

```python
file.head()
```

```python
file["flower type"] = file['target'].apply(lambda x : iris.target_names[x])
```

```python
file.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(15, 12))
df0 = file[file['target'] == 0]
df1 = file[file['target'] == 1]
df2 = file[file['target'] == 2]

ax0.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='blue')
ax0.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='red')
ax0.set(
    title="setosa VS vericolor comparison between sepal",
    xlabel="sepal length", 
    ylabel='sepal width'
);

ax1.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue')
ax1.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='red')
ax1.set(
    title="vericolor vs virginica comparison between sepal",
    xlabel="sepal length", 
    ylabel='sepal width'
);

ax2.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='blue')
ax2.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='red')
ax2.set(
    title="setosa vs virginica comparison between sepal",
    xlabel="sepal length", 
    ylabel='sepal width'
);

ax3.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='blue')
ax3.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='red')
ax3.set(
    title="setosa VS vericolor comparison between sepal",
    xlabel="petal length", 
    ylabel='petal width'
);
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

X = file.drop(['target', 'flower type'], axis=1)
y = file['flower type']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)

model = SVC()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

X = file.drop(['target', 'flower type'], axis=1)
y = file['flower type']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)

model = SVC(kernel='poly')
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
