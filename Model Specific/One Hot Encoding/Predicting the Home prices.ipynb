```python
import pandas as pd
```

```python
file = pd.read_csv('homeprices.csv')
```

```python
file.head()
```

```python
dummies = pd.get_dummies(file["town"])
```

```python
file = pd.concat([file, dummies], axis=1)
```

```python
file
```

```python
file.drop('town', axis=1, inplace=True)
```

```python
file
```

```python
from sklearn.ensemble import RandomForestRegressor

X = file.drop("price", axis = 1)
y = file["price"]

model = RandomForestRegressor()

model.fit(X, y)
```

```python
model.predict([[3000, 1, 0, 0]])
```

```python
import pickle

with open("modelFormedAfterdummies.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
car_prices = pd.read_csv('carprices.csv')
```

```python
car_prices
```

```python
car_prices.isna().sum()
```

```python
dummies = pd.get_dummies(car_prices['Car Model'])
```

```python
car_prices = pd.concat([car_prices, dummies], axis=1)
```

```python
car_prices
```

```python
car_prices.drop("Car Model", axis = 1, inplace = True)
```

```python
car_prices
```

```python
X = car_prices.drop('Sell Price($)', axis=1)
y = car_prices["Sell Price($)"]
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X, y)
```

```python
model.predict([[69000, 6,0,1,0]])
```

```python
import pickle
with open("car sales model.pkl", "wb") as f:
    pickle.dump(model, f)
```
